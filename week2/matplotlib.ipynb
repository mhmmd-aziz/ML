## line plot
### digunakan untuk perubahan nilai sewaktu waktu kayak trading dan time series

```python
x = [1, 4, 6, 7, 8]
y = [3, 6, 8, 9, 4]
plt.figure(figsize=(12, 6))
plt.plot(x, y)
plt.xlabel("sumbu x")
plt.ylabel("sumbu y")
plt.title("grafik aziz")
plt.grid(True)
plt.show()
```

```python
x1 = [12, 60, 80, 67, 88]
x2 = [52, 60, 88, 67, 98]
x3 = [82, 60, 83, 67, 68]
plt.plot(x1, x3, marker="o", color="red", linestyle='-', label='mtk')
plt.fill(x1, x3, color='red', alpha=0.1, label='area terarsir')
plt.plot(x2, x3, marker="o", color="green", linestyle='-', label='coding')

#detail element
plt.title('grafik nilai ujian fadhil')
plt.xlabel("sumbu x")
plt.ylabel("sumbu y")
plt.legend(loc=('upper right'))
plt.grid(True)
plt.savefig('.\week2')
plt.show()
```

# barplot

```python
kategori = ["python", 'c++', 'php']
jumlahpengguna = [100, 50, 47]
plt.figure(figsize=(12, 7))
plt.bar(kategori, jumlahpengguna, label="trend bahasa pemrograman")
plt.title('pengguna bahasa pemrograman')

plt.xlabel("kategori bahasa pemroraman")
plt.ylabel("jumlah penggunaan bahasa pemrograman")
plt.grid(True)
plt.legend(loc=("upper right"))
plt.show()
plt.savefig(".\week2")
```

```python
event = ['satu', 'dua', 'tiga'] 
laki = [34, 67, 89]
perempuan = [100, 50, 57]
x=np.arange(len(event))
barwidth = 0.35

plt.figure(figsize=(12, 6))
plt.bar(x - barwidth/2, laki, label="laki-laki")
plt.bar(x + barwidth/2, perempuan, label="perempuan")
plt.title('pengguna bahasa pemrograman')

plt.xlabel("kategori bahasa pemroraman")
plt.ylabel("jumlah penggunaan bahasa pemrograman")
plt.grid(axis='y')
plt.legend(loc=("upper right"))
plt.tight_layout()
plt.show()
plt.savefig(".\week2")
```

# scatter plot
### digunakan untuk melihat hubungan antara variabel

```python
umur = [18, 21, 25, 30, 35, 40, 45, 50]
penghasilan = [1.5, 2.2, 2.8, 4.0, 4.2, 4.5, 4.7, 4.9]
plt.Figure(figsize=(8, 5))
plt.scatter(penghasilan, umur)
plt.title('hubungan antara gaji dan umur')
plt.xlabel('penghasilan')
plt.ylabel('umur')
plt.grid(True)
#plt.legend()
plt.tight_layout()
plt.show
```

```python
fisika_A = [75, 80, 85, 90, 78]
matematika_A = [70 ,85, 82, 90, 99]

fisika_B = [60, 65, 61, 20, 90]
matematika_B = [65, 62, 70, 80, 11]

fisika_C = [70, 60, 61, 20, 90]
matematika_C = [65, 50, 50, 80, 11]

plt.Figure(figsize=(8, 5))
plt.scatter(matematika_A, fisika_A, marker='o', color='red', label='nilai kelas a')
plt.scatter(matematika_B, fisika_B, marker='o', color='green', label='nilai kelas b')
plt.scatter(matematika_C, fisika_C, marker='*', color='blue', label='nilai kelas c')

plt.title('nilai mtk fisika kelas a, b dan c')
plt.xlabel('matematika')
plt.ylabel('fisika')
plt.grid(True)
plt.legend(loc='lower left')
plt.tight_layout()
plt.savefig('..\week2scater.png.')
plt.show
```

## subsplot

```python
hari = ['Sen', 'Sel', 'Rab', 'Kam', 'Jum', 'Sab', 'Min']
followers = [120, 130, 145, 160, 175, 190, 210]
views = [300, 400, 350, 500, 600, 550, 700]
likes = [100, 120, 110, 160, 200, 180, 250]
comments = [20, 25, 22, 35, 40, 38, 50]
engagement_rate = [0.4, 0.36, 0.38, 0.39, 0.4, 0.39, 0.43]
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

axs[0,0].plot(hari, followers, marker='o')
axs[0,0].set_title('pertumbuhan followers')
axs[0,0].set_ylabel('jumlah_followers')
axs[0,0].grid(alpha=0.3)

axs[0,1].bar(hari, views, color='skyblue')
axs[0,1].set_title('pertumbuhan views')
axs[0,1].set_ylabel('jumlah_views')
axs[0,1].grid(alpha=0.3)

axs[1,0].plot(hari, likes, marker='o')
axs[1,0].set_title('pertumbuhan likes')
axs[1,0].set_ylabel('jumlah_likes')
axs[1,0].grid(alpha=0.3)

axs[1,1].scatter(hari, engagement_rate, color='orange', marker='o')
axs[1,1].set_title('pertumbuhan engagement_rate')
axs[1,1].set_ylabel('jumlah_engagement_rate')
axs[1,1].grid(alpha=0.3)
axs[1,1].set_ylim(0, 0.5) # memaksa sumbu-y hanya menampilkan rentang dari 0 sampai 0.5.


fig.suptitle('Trafic Media Sosial', fontsize=32)
plt.tight_layout()
plt.savefig('..\week2\media sosial.png')
```

## corelation matrix

```python
df.head()
```

```python
df.dtypes
```

```python
numerical_df = df[['Age', 'Quantity', 'Price per Unit', 'Total Amount']]
```

```python
import seaborn as sns
```

```python
correlation_matrix = numerical_df.corr()
correlation_matrix
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True, fmt='.2f', linewidths=0.5, cbar_kws={'label':'koofisien korelasi'})
plt.title('matrix corelation', fontsize=32)
plt.xticks(rotation=0)
plt.yticks(rotation=90)
plt.yticks()
plt.tight_layout()
plt.show()
```
