import pandas as pd

df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})

# Jumlah per kolom (axis=0, default)
print(df.sum(axis=0))
# Output:
# A     6
# B    15

# Jumlah per baris (axis=1)
print(df.sum(axis=1))
# Output:
# 0     5   (1+4)
# 1     7   (2+5)
# 2     9   (3+6)

## karena pas di jadikan fram jadi a dan b bukan a b keawah kayak di kode

import pandas as pd

df = pd.DataFrame({
    'A': [1,2,3],
    'B': [4,5,6],
    'C': [7,8,9]
})

# Hapus baris dengan index 0
df_drop_row = df.drop(0, axis=0)
print(df_drop_row)
# Output: baris pertama hilang

# Hapus kolom 'B'
df_drop_col = df.drop('B', axis=1)
print(df_drop_col)
# Output: kolom B hilang
