## remove outlier atau anomali

```python
def remove_outlier_iqr(df, columns, x):
    df_clean = df.copy()
    for col in columns:
        q1 = df_clean[col].quantile(0.25)
        q3 = df_clean[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - x * iqr #1.5
        upper_bound = q3 + x * iqr #1.5
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)] 
    return df_clean
```

```python
df_no_outliers = remove_outlier_iqr(dataframenew, numerical_columns, 2)
```

```python
df_no_outliers
```

```python
from sklearn.preprocessing import MinMaxScaler
df_min_max_scaler = df_no_outliers.copy()
min_max_scaler = MinMaxScaler()
df_min_max_scaler[numerical_columns] = min_max_scaler.fit_transform(df_no_outliers[numerical_columns])
```

```python
df_min_max_scaler.describe()
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```python
df_standardscaler = df_no_outliers.copy()
df_standardscaler[numerical_columns] = std_scaler.fit_transform(df_no_outliers[numerical_columns])
```

```python
df_standardscaler.describe()
```

```python
from sklearn.preprocessing import RobustScaler
robust_scaler = RobustScaler()
df_robust_sclaer = df_no_outliers.copy()
df_robust_sclaer[numerical_columns] = robust_scaler.fit_transform(df_no_outliers[numerical_columns])
```
